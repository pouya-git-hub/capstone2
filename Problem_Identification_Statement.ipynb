Predict Electricity Consumption


Problem Statement:
In order to evaluate the future budget and resources, the Department of Energy wants to have a reliable prediction of the state’s electricity consumption for the next 6 years from 2019 to 2024.

Context:
The Department of Energy is seeking to expand the state’s power plants facilities in order to cover future consumption capacity. Therefore, the department has requested us to predict electricity consumption for the next 6 years, from 2019 to 2024.  

Criteria for Success:
Predicting the state electricity consumption volume for the next 6 consecutive years, from 2019 to 2024.

Scope of Solution Space:
As the first step, the provided data which includes electricity consumption between the years 1985 to 2018 need to be cleaned, and visualized. Then by using a variety of Machine Learning models, we will provide a reliable prediction about future electricity consumption to the Department of Energy.  

Constraints:
The dataset we have includes date and volume of consumption. However, it does not have any information about the state population and users’ energy consumption behaviour changes over the past 33 years.
The state's current population growth is unknown.
There is no separation between residential and industrial consumption.

Stakeholders:
The Department of Energy

Data Sources:
The state’s past 33 years of electricity consumption history in the form of a spreadsheet. The consumption history starts from 1985 and ends in 2018.
The data is picked from Kaggle.com. Here is a link to the dataset:
https://www.kaggle.com/code/nageshsingh/predict-electricity-consumption
